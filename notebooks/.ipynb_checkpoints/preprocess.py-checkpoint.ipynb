```python
import pandas as pd
```

```python
data = pd.read_csv('../datasets/jadili/negative2.csv')
```

```python

char_to_remove = set(['?', '!', '$', ')','(', '^','`', '*', '-', '_', '+', '}', '{', ':', ';', '<', '>', '/', '\\', '|'])

data['tweet']=data['tweet'].map(lambda tweet: ''.join([t for t in tweet if t not in char_to_remove]))
```

```python
import re
# Remove Mentions
data['tweet']=data['tweet'].map(lambda tweet: re.sub("@\w+", "", tweet) )

# Remove Tags
data['tweet']=data['tweet'].map(lambda tweet: re.sub("#\w+", "", tweet) )
# 5900
# 17700

# 5000, 5000
```
